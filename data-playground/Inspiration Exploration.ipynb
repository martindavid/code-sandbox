```python
MIN_SIMILARITY = 0.5
MIN_VALUABLE_SCORE = 0.7

filter_condition = (
    (data['wmd'] >= MIN_SIMILARITY),
    (data['wmd'] < MIN_SIMILARITY) &
    (data['wmd_name'] < MIN_SIMILARITY) &
    (data['valuable_score'] > MIN_VALUABLE_SCORE)
)
filter_choices = ['Incore', 'Yes']
```

```python
data['accept'] = np.select(
    condlist=filter_condition, choicelist=filter_choices, default="No")
data.loc[(data['description'].duplicated()) & (
    data['accept'] == 'Yes'), 'accept'] = "No"
```
